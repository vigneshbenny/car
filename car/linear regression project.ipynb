```python
df[['CarName','fueltype', 'aspiration','doornumber', 'carbody', 'drivewheel', 'enginelocation','fuelsystem','enginetype','cylindernumber']] = df[[ 'CarName','fueltype', 'aspiration','doornumber', 'carbody', 'drivewheel', 'enginelocation','fuelsystem','enginetype','cylindernumber']].apply(LabelEncoder().fit_transform)
```

```python
import pandas as pd
import numpy as np


# Function to remove outliers using Z-score
def remove_outliers(df, columns, threshold=3):
    z_scores = np.abs((df[columns] - df[columns].mean()) / df[columns].std())
    df_no_outliers = df[(z_scores < threshold).all(axis=1)]
    return df_no_outliers

# Specify columns for outlier removal
columns_to_check = ['drivewheel','wheelbase','carlength','carwidth','curbweight','enginesize',
'fuelsystem'
,'boreratio',
'horsepower',
'citympg',
'highwaympg']

# Remove outliers
df_no_outliers = remove_outliers(df, columns_to_check)

print("Original DataFrame:")
print(df)
print("\nDataFrame without outliers:")
print(df_no_outliers)
```

```python
x=df_no_outliers[['CarName', 'fueltype', 
       'doornumber', 'carbody', 'drivewheel', 'enginelocation', 'wheelbase',
         'curbweight', 'enginetype',
        'enginesize','boreratio', 'stroke',
        'horsepower', 'peakrpm', 'citympg', 'highwaympg']]
y=df_no_outliers["price"]
x
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model=LinearRegression(learning_rate=0.001, n_iterations=1000)
model.fit(X_train,y_train)
```

```python
y_p=model.predict(X_test)
y_p
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np



# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_p)
print("Mean Absolute Error (MAE):", mae)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_p)
print("Mean Squared Error (MSE):", mse)

# Calculate R-squared (R^2)
r2 = r2_score(y_test, y_p)
print("R-squared (R^2):", r2)
```
